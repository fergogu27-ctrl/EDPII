# **CAMINATA ALEATORIA 1D**

En este ejercicio iniciamos en 0 en una recta y hay una probabilidad p de moverse  1 a la derecha y una probabilidad q de moverse -1 a la izquierda  y se cumpla que p+q=1. Lo que proponemos a continucación y graficamos dicha caminata.




```python
import numpy as np
import random as r  #libreria que nos ayuda a generar numeros aleatorios
import matplotlib.pyplot as plt  #libreria que nos ayuda a graficar
```

En la clase pudimos observar una recta donde podemos definir de la siguiente manera.


*   Posición inicial $x_0 = 0$


*   Incrementos aleatorios
$\ \ +1 \ con \ probabilidad \ p \\
-1 con probabilidad q=1-p$







```python
def caminata(a,p):
  posicion = 0
  trayectoria = [posicion]

  for i in range(a):
        r = np.random.rand()  # número aleatorio entre (0,1)

        if r < p:
            posicion = posicion + 1
        else:
            posicion = posicion - 1

        trayectoria.append(posicion)

  return trayectoria

```

```python
a = 100      # numero de pasos
p = 0.5      # probabilidad de ir a la derecha podemos decir que hay un 50% de porbabilidad

trayectoria = caminata(a, p)

plt.plot(trayectoria)
plt.xlabel("n (paso)")
plt.ylabel("X_n (posición)")
plt.title(f"Caminata aleatoria 1D con p = {p}")
plt.grid(True)
plt.show()
```

En cada paso de la caminata se genera un número aleatorio entre (0,1).
Donde si $r < p$ sabemos que la recta se movio una unidad a la derecha, en lo contrario, $r>p$ se movio a la izquierda.

Algunas preguntas que podemos ejecutar

Despues de n pasos ¿Cuál es la posición promedio de los movimientos?

```python
def promedio_posicion(a, p, N):
    posiciones_finales = []
    for i in range(N):
        trayectoria = caminata(a, p)
        posiciones_finales.append(trayectoria[-1])
    return np.mean(posiciones_finales)

```

```python
a = 100
p = 0.5
N = 1000

print("Promedio simulado:", promedio_posicion(a, p, N))
print("Promedio teórico :", a*(2*p - 1))
```

¿La caminata regresa al cero? Hacemos una lista para despues comparar si es que regreso al 0

```python
def regresa_a_cero(a, p):
    trayectoria = caminata(a, p)
    return 0 in trayectoria[1:]  # ignora X0
```

```python
a = 1000
p = 0.6
N = 500

conteo = sum(regresa_a_cero(a, p) for _ in range(N))
print("Fracción que regresa al cero:", conteo / N)
```

¿Cuanto se alejo respecto al origen? Podemos calcular por medio del valor absoluto

```python
def distancia_maxima(a, p):
    trayectoria = caminata(a, p)
    return max(abs(x) for x in trayectoria)
```

```python
a = 1000
p = 0.5

print("Distancia máxima al origen:", distancia_maxima(a, p))
```
