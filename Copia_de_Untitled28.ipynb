Algoritmo para simular una caminata de Markov

Entrada:  
$\alpha$ = Distribucion inicial

P La matriz

n el numero de pasos

Queremos que de una sucesion de n pasos del proceso, debemos decir quien es s
Se inicia generando el valor inicial con una variable que tenega la misma distribucion o sea que sea la probabilidad =1/4

 Tenemos que imprimir como la salida la sucesion en forma de una lista

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
alpha = [1/4, 1/4, 1/4, 1/4]      #Es nuestra distribuión inicial
S = [0, 1, 2, 3]                  #Numero de estados

P = np.array([                    #Matriz de probabilidad (de transición P)
    [1/4, 0,   1/4, 2/4],
    [2/4, 1/4, 0,   1/4],
    [1/4, 2/4, 1/4, 0  ],
    [0,   1/4, 2/4, 1/4]
])
```

```python
def CM(alpha, P, n ):         #Establecemos la función para la cadena de Markov
  x0=random.choices(S, weights=alpha)[0]    #Generamos el estado inicial x0 ~ alpha
  trayectoria = [x0]

  for i in range(n):                        #Generamos los siguienetes x1, x2, ..., xn
    estado0 = trayectoria[-1]               #Tomamos el estado actual
    estado1=random.choices(S, weights=P[estado0])[0]
    trayectoria.append(estado1)

  return trayectoria
```

```python
n=100         #Número de pasos
```

```python
sucesionf = CM(alpha, P, n)        #Hacemos la simulacion teniendo todos los datos
print(sucesionf)                      #Cada número indica el estado en cada paso
```

```python
probabilidad0 = sucesionf.count(0)/n
print(probabilidad0)
probabilidad1 = sucesionf.count(1)/n
print(probabilidad1)
probabilidad2 = sucesionf.count(2)/n
print(probabilidad2)
probabilidad3 = sucesionf.count(3)/n
print(probabilidad3)
```

```python
colores_estados = {
    0: "rojo",
    1: "naranja",
    2: "verde",
    3: "azul"
}
```

```python
print("\n Sucesion colores:")
print([colores_estados[x] for x in sucesionf])
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(n+1), sucesionf, marker='o')
plt.xlabel("Paso")
plt.ylabel("Estado")
plt.title("Trayectoria de la cadena de Markov")
plt.yticks([0,1,2,3], ["rojo", "naranja", "verde", "azul"])
plt.grid(True)
plt.show()
```
