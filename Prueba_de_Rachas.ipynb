# **Prueba de Rachas**
Prueba de Independencia

Prueba estadisticas para los números pseudoaleatorios.


Para esto plantearemos la prueba de Independecia mejor conocida como prueba de rachas  donde prueba la hipótesis de la distribución acumulada de una variable aleatoria x en $F_0(x)$.

En esta prueba contamos el número de rachas y calculamos el valor esperado, la varianza y por ultimo hacer una transformación de estos datos para que obtengamos la distribución normal estandar $F(x)$.

Establecemos una hipotesis nula $H_0$ la cual se debe de probar y una hipotesis alternativa  $H_a$ (hipotesis de investigación) describira lo que se concluirá si se rechaza la hipotesis nula.

Para esta prueba se considera la siguienete hipotesis nula:


Los datos proviene de :

Hipotesis nula:   Los números de la muestra ${x_i}$ son independientes



Hipotesis alternativa: Los números de la muestra ${x_i}$ no son independientes


```python
import numpy as np


data_m = np.array([   #Hacemos el arreglo para que pueda leer de manera correcta las columnas
0.78961, 0.05230, 0.10699, 0.55877, 0.14151,
0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
0.80548, 0.82654, 0.29453, 0.20852, 0.42989,
0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
0.89898, 0.57880, 0.67621, 0.05010, 0.00121,
0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
0.38618, 0.76910, 0.68334, 0.55170, 0.10850,
0.79882, 0.45679, 0.21631, 0.87616, 0.55743,
0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
0.29931, 0.30861, 0.83358, 0.51781, 0.03272,
0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
0.24000, 0.65559, 0.38507, 0.90829, 0.94187,
0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
0.54325, 0.62400, 0.09133, 0.41678, 0.33954,
0.58244, 0.85853, 0.86752, 0.38729, 0.15506,
0.23949, 0.53559, 0.33381, 0.49883, 0.75103,
0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
0.19147, 0.40644, 0.08128, 0.73435, 0.22724,
0.22287, 0.07281, 0.64183, 0.44267, 0.72102
],  dtype=float)
```

En esta prueba puede ser descrita por los siguientes pasos.
1. Determinar las secuencias de 0 y 1 (+ y -) así:


*   Si $x_i\leq x_{i-1}$ escribir 0 (-)
*   Si $x_i > x_{i-1}$ escribir 1 (+)

    La secuencia S contiene n-1 números:

    S={-,-,+,-,+,+,+,-,-,+,-,+}

    Contamos como racha cada que se cambio de signo en este ejemplo tendemos 8 rachas.

2. Contar el número de rachas observandolas. Una racha viene dada por el número de ceros o unos consecutivos que la forman y se cuenta otra cuando esta cambia.

    NOTA: Es una caminata aleatoria de procesos estocasticos.

3. Calcular el valor esperado y la varianza del número de rachas, así como el estocastico $Z_0$ mediante:
*   Media:
$$\mu_{C_o} = \frac{2n-1}{3}$$
*   Varianza:
$$\sigma^2_{C_0} = \frac{16n-29}{90}$$

*   Estocastico:
$$z_0 = \frac{C_n-\mu_{C0}}{\sigma_0 }$$

```python
M = data_m.reshape(20, 5)  # Tenemos 5 filas


data = M.flatten(order='F')   # 'F' = Fortran (por columnas) las datos se ponen en forma de columna para que se pueda realizar los pasos.
n = len(data)

n=100   #Son nuestros n datos de la lista del ejercicio
```

Cuando verifiquemos que el programa pueda leer los datos en forma de columna aplicamos lo pasos anteriormente mencionados.

```python
# Paso 1
signos = np.where(data[1:] > data[:-1], 1, 0)  # 1 = '+', 0 = '-' #Checamos la secuencia de signos + si sube y - si no sube
print("S:",signos)

# Paso 2
C0 = 1 + np.sum(signos[1:] != signos[:-1])    #Contamos las rachas , indicamos que cada cambio de signo se concidera una racha


signos = []      #Contruimos el conteo de signos para saber cuantas rachas tenemos

for i in range(1, n):
    if data[i] > data[i-1]:
        signos.append("+")
    else:
        signos.append("-")

rachas = 1  #Se cuentan las rachas
for i in range(1, len(signos)):
    if signos[i] != signos[i-1]:
        rachas += 1

print("Número de rachas:", rachas)
```

El anterior fue un poco complicado, con la lectura y los datos y verificar que las rachas fueran las correctas ya teniendo estos datos podemos aplicar la formulas.

```python
# Paso 3
mu = (2*n - 1) / 3        #Media teorica
var = (16*n - 29) / 90    #Varianza teorica
sigma = np.sqrt(var)      #Desviación estandar

z0 = (rachas - mu) / sigma    # estadistico z0
```

Dado que $C_0$ sigue una distribución normal de media $\mu_{C0}$ y una varianza $\sigma^2_{C0}$ y el estadistico $Z_0$ una distribución normal [0,1).

El intervalo de aceptación es:
*   $$-Z_{\frac{\alpha}{2}}\le Z_0\le Z_{\frac{\alpha}{2}} $$

Sabemos que por la decisión al 95% que es la que hemos ocupado $\alpha=0.05$ por lo tanto sabemos que el valor critico es:  $$Z_{\frac{\alpha}{2}}=1.96$$

```python
print("Numero de rachas:", rachas)
print("Media teorica:", round(mu,4))
print("Varianza teorica:", round(var,4))
print("Desviacion estandar:", round(sigma,4))
print("Z0:", round(z0,4))

if abs(z0) <= 1.96:
    print("No se rechaza H0: los números de la muestra son independientes")
else:
    print("Se rechaza H0: los números de la muestra no son independientes")
```

Esto nos indica que los datos de la muestra estan "revueltos" ya que si estuvieran en orden de mayor a menor o de menor a mayor el cambio de signo seria el mismo siempre y nos saldria que solo tendriamos una racha y al realizar los calculos la hipotesis seria rechazada. Es por esto que necesariamente deben estar revueltos y nos indica que son pseudoaleatorios.
