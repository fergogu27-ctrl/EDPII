***Valores y vectores propios***

Para encontrar los valores propios de una matriz utilizamos eigenvals.eigenvals devuelve un diccionario de pares.

```python
M = Matrix([[3, -2,  4, -2], [5,  3, -3, -2], [5, -2,  2, -2], [5, -2, -3,  3]])
M
```

```python
M.eigenvals()
```

Esto significa que M tiene valores propios -2, 3 y 5, y que los valores propios -2 y 3 tienen multiplicidad algebraica 1 y el valor propio 5 tiene multiplicidad algebraica 2

Para encontar los valores propios de una matriz, utilizamos eigenvects.eigenvects donde devuelve una lista de tuplas de la forma (eigenvalue, algebraic_multiplicity, [eigenvectors])

```python
M.eigenvects()
```

Esto nos muestra que, por ejemplo, el valor propio 5 también tiene multiplicidad geométrica 2, ya que tiene dos vectores propios. Dado que las multiplicidades algebraicas y geométricas son las mismas para todos los valores propios, M es diagonalizable

```python
for lam, mult, vecs in M.eigenvects():
    print("Valor propio:", lam, " multiplicidad:", mult)
    for i, vec in enumerate(vecs, start=1):
        print(f"  v{i} =", vec)
```

**Libreria Networkx**

Esta libreria nos permite realizar la creación, manioulación y analisis de grafos y/o redes complejas. Nos ayuda al dibujo de redes.
Creando nodos y aristas visualizamos grafos y aplicaciones de funciones basicas para que nos de una lista de nodos que componen el grafico, las rutas ó artistas que existen así como el número de nodos. Si la libreria no esta instalada se instala con la siguiente linea.

Para importar la libreria para trabajar con grafos

```python
import networkx as nx
```

Primero vemos como creamos un gráfico sin nodos ni artistas

Un nodo Graph es una colección de nodos (vertices) junto con pares de nodos identificados. Los nodos pueden ser cualquier objeto hash.

```python
G = nx.Graph()
```

Hay tipos de graficos:


*   Graph(): Función para crear un grafo no dirigido y no permite aristas paralelas.
*   DiGraph(): Función para crear un grafo dirigido,proporciona métodos y propiedades adicionales específicos para bordes dirigidos.
*   MultiGraph(): Función para crear un grafo no dirigido con multiples aristas entre nodos.
*   MultiDiGrap(): Función para crear un grafo dirigido con múltiples aristas entre nodos.







***Nodos***

Podemos agregar un nodo a la vez

```python
G.add_node(1)
nx.draw_networkx(G)
```

O podemos agregar nodos desde cualquier contenedor iterable, como una lista .

```python
G.add_nodes_from([2, 3])
nx.draw_networkx(G)
```

Tambien puede agregar nodos junto con atributos de nodo si su contenedor produce 2 tuplas del formato (node, node_attribute_dict)

```python
G.add_nodes_from([(4, {"color": 'red'}), (5, {"color": "green"})])
nx.draw_networkx(G)
```

***Aristas***

Los aristas tambien se puede llamar bordes donde se puede cultivar añadiendo un borde a la vez

```python
G.add_edge(1, 2)      #Agregamos una arista y borde entre 1 y 2
nx.draw_networkx(G)
```

```python
e = (2,3)     #Agregramos un borde utilizando tupla
G.add_edge(*e)
nx.draw_networkx(G)
```

```python
G.add_edges_from([(1, 3), (3, 4), (4, 5)])    #Agregamos múltiples aristas con los datos establecidos de borde
nx.draw_networkx(G)
```

Podemos examinar los nodos y las aristas , con cuatro propiedades básica de grafos que facilitan la generación de informes:


*   G.nodes: Función que devuelve el número de nodos
*   G.edges: Función que devuelve el número de las conexiones
*   G.adj[nodo]: Función que muestra los vecinos de un nodo.
*   G.degree[nodo]: Función que muestra cuántas aristas tiene ese nodo.
*   nx.number_of_nodes(): Está función me devuelve la cantidad de nodos.
*   nx.number_of_edges() # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras).
*   nx.shortest_path(G, source, target, weight=None): Función que me da el camino mas corto del Grafo.

```python
# Listar nodos y aristas
print("Nodos:", G.nodes)
print("Aristas:", G.edges)

# Ver nodos adyacentes a 1
print("Vecinos de 1:", list(G.adj[1]))

# Ver el grado de un nodo (cuántas conexiones tiene)
print("Grado de 1:", G.degree[1])

print("Cantidad de nodos", nx.number_of_nodes(G)) #Está función me devuelve la cantidad de nodos

print("Cantidad de Aristas",nx.number_of_edges(G)) # Está función me devuelve la cantidad de Aristas (Arcos, Ramas o Ligaduras)

print("El camino mas corto", nx.shortest_path(G, 1, 5, weight=None))
```

Para eliminar Nodos y Aristas
Se puede borrar los nodos y tadas sus aritas
Tambien se borra solo ciertas conexiones sin eliminar el nodo.


*   remove_node(nodo): Función que eliminan un nodo
*   G.remove_nodes_from([a,b]): Función que elimina varios nodos.
*   G.remove_edge(a,b): Función que elimina una arista.



```python
G.remove_node(3)    # Eliminar un nodo 1, 2, 3, 4, 5
nx.draw_networkx(G)
```

```python
nodes_to_remove = [node for node in [1, 2] if node in G]    # Eliminar varios nodos
G.remove_nodes_from(nodes_to_remove)
nx.draw_networkx(G)
```

```python
if G.has_edge(1, 3):    # Eliminar una arista
    G.remove_edge(1, 3)

nx.draw_networkx(G)
```

En lugar de agregar nodos y aristas paso a paso, o uno por uno se puede crear grafos directamente:


*   Graph([(a, b), (b, c), ...]): Función que construye un grafo directamente con las conexiones.
*   Graph(diccionario): Función que usa un diccionario de adyacencia, donde cada clave es un nodo y su valor es la lista de vecinos.



```python
H = nx.Graph([(0, 1), (1, 2), (2, 3)])    # Grafo a partir de una lista de aristas

nx.draw_networkx(H)
```

```python
adj_dict = {0: [1, 2], 1: [0, 2], 2: [0, 1]}    # Grafo desde diccionario de adyacencia
J = nx.Graph(adj_dict)
nx.draw_networkx(J) #Funcion que dibuja el grafo.
```

NetworkX tiene funciones sencilla para dibujar, como las siguientes:


*   nx.draw: Dibuja el grafo.
*   with_labels=True: hace que los nodos se muestren con su nombre.

Hay que aclarar que esta libreria es solo de uso básico, si se necesita para visualizaciones más avanzadas se recomienda usar otras herramientas como: Gephi, Cytoscape, Graphviz


```python
import matplotlib.pyplot as plt

nx.draw(H, with_labels=True)
plt.show()
```
