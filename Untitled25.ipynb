```python
import numpy as np
import random as r  #libreria que nos ayuda a generar numeros aleatorios
import matplotlib.pyplot as plt  #libreria que nos ayuda a graficar

def caminata(a,p):
  posicion = 20
  trayectoria = [posicion]

  while posicion != 0 and posicion != 60:

        r = np.random.rand()  # número aleatorio entre (0,1)

        if r < p:
            posicion = posicion + 1
        else:
            posicion = posicion - 1

        trayectoria.append(posicion)


  return trayectoria

a = 1000      # numero de pasos
p = 0.5      # probabilidad de ir a la derecha podemos decir que hay un 50% de porbabilidad

trayectoria = caminata(a, p)

plt.plot(trayectoria)
plt.xlabel("n (paso)")
plt.ylabel("X_n (posición)")
plt.title(f"Caminata aleatoria 1D con p = {p}")
plt.grid(True)
plt.show()

def regresa_a_cero(a, p):
    trayectoria = caminata(a, p)
    return 0 in trayectoria

a = 1000
p = 0.5

N = 1000


exitos = sum(regresa_a_cero(a, p) for _ in range(N))#trayevtoria

probabilidad = exitos / N
print("Probabilidad de llegar a cero:", probabilidad)
```
